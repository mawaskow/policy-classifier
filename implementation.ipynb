```python
from populate_corpora.pdfs_to_jsons import pdfs_to_txt_dct
import nltk
from populate_corpora.data_cleaning import get_clean_text_sents
EN_TOKENIZER = nltk.data.load("tokenizers/punkt/english.pickle") # need tokenizer for our text cleaning
import torch
from tqdm import tqdm
from sklearn import svm
from sentence_transformers import SentenceTransformer
```

```python
pdf_dict = pdfs_to_txt_dct("C:/Users/allie/Documents/PhD/to_analyse")
```

```python
clean_sents= get_clean_text_sents(pdf_dict, EN_TOKENIZER)
```

```python
sentences = [
    "The authority can revise the delegated act every five years.",
    "The scheme will subsidise purchases of eco-friendly farm equipment.",
    "Farmers will be able to avail of expert assistance in the uptake of new technologies."
]
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tok = AutoTokenizer.from_pretrained("mawaskow/inc_sent_cls_mc")
model = AutoModelForSequenceClassification.from_pretrained("mawaskow/inc_sent_cls_mc")

text = sentences[1]
inputs = tok(text, return_tensors="pt")

with torch.no_grad():
    logits = model(**inputs).logits

pred = torch.argmax(logits, dim=-1).item()
print(model.config.id2label[pred])
```

```python
bn_model = SentenceTransformer("mawaskow/inc_sent_cls_bn")
```

```python
mc_model = SentenceTransformer("mawaskow/inc_sent_cls_mc")
```

```python
import numpy as np
train_stacked = np.vstack([bn_model.encode(sent) for sent in tqdm(sentences)])#bn_df["train"]["text"])])
train_embs = [torch.from_numpy(element).reshape((1, element.shape[0])) for element in train_stacked]
use_stacked = np.vstack([bn_model.encode(sent) for sent in tqdm(clean_sents)])
use_embs = [torch.from_numpy(element).reshape((1, element.shape[0])) for element in use_stacked]
clf = svm.SVC(gamma=0.001, C=100., random_state=9)
clf.fit(np.vstack(train_embs), bn_df["train"]["label"])
```

```python
prds_lst = [clf.predict(sent_emb)[0] for sent_emb in use_embs]
```

```python
incs = []
for sent, label in zip(clean_sents, prds_lst):
    if label==1:
        incs.append(sent)
```

```python
id2label = {
    0: "Fine",
    1: "Supplies",
    2: "Technical_assistance",
    3: "Tax_deduction",
    4: "Credit",
    5: "Direct_payment"
}
```

```python
train_stacked = np.vstack([bn_model.encode(sent) for sent in tqdm(mc_df["train"]["text"])])
train_embs = [torch.from_numpy(element).reshape((1, element.shape[0])) for element in train_stacked]
use_stacked = np.vstack([bn_model.encode(sent) for sent in tqdm(incs)])
use_embs = [torch.from_numpy(element).reshape((1, element.shape[0])) for element in use_stacked]
clf = svm.SVC(gamma=0.001, C=100., random_state=9)
clf.fit(np.vstack(train_embs), mc_df["train"]["label"])
prds_lst = [clf.predict(sent_emb)[0] for sent_emb in use_embs]
```

```python
results = {
    0:[],
    1:[],
    2:[],
    3:[],
    4:[],
    5:[],
}
```

```python
for sent, label in zip(incs, prds_lst):
    results[label].append(sent)
```

```python
for id in list(id2label):
    if results[id]:
        print(f"\n{id2label[id]}")
        for result in results[id]:
            print(result)
```
